Probability of full payment and  min_payment_amt variables have outliers

```python
#Treating outliers

def find_whiskers(col):
    sorted(col)
    Q1 , Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range , upper_range

for col in ['probability_of_full_payment', 'min_payment_amt']:
    lwrange , uprange = find_whiskers(data[col])
    data[col] = np.where(data[col] > uprange , uprange , data[col])
    data[col] = np.where(data[col] < lwrange , lwrange , data[col])
```
