```python
#Outlier Treatment
def find_whiskers(col):
    sorted(col)
    Q1 , Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range , upper_range

for col in ['AgentBonus', 'Age','CustTenure','ExistingProdType','MonthlyIncome','ExistingPolicyTenure','SumAssured','LastMonthCalls']:
    lwrange , uprange = find_whiskers(df[col])
    df[col] = np.where(df[col] > uprange , uprange , df[col])
    df[col] = np.where(df[col] < lwrange , lwrange , df[col])
```
